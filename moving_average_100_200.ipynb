```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("TCS.NS.csv")
```

```python
print(data)
```

```python
ma100=data.Close.rolling(100).mean()
ma200=data.Close.rolling(200).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.Close)
plt.plot(ma100, 'r')
plt.plot(ma200, 'g')
plt.grid()
plt.xlabel('Year ----->')
plt.ylabel('Cost in  ----->')
plt.legend(["Closing Value", "100 days moving avg", "200 days moving avg"])  # show legends
plt.show()
```

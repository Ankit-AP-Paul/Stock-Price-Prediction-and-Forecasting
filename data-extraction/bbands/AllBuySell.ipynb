```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:\\Users\\User\\Downloads\\abc\\ADANIGREEN.NS.csv",index_col='Date')
```

```python
df.head()
```

```python
df['30_MA_Close']=df['Close'].rolling(window=30).mean()
df['20_STD_Close']=df['Close'].rolling(window=20).std()
```

```python
df.head(35)
```

```python
df[['Close','30_MA_Close']].plot(figsize=(10,5))
```

```python
df['Upper']=df['30_MA_Close']+2*df['20_STD_Close']
df['Lower']=df['30_MA_Close']-2*df['20_STD_Close']
```

```python
df.head(35)
```

```python
df[['Close','30_MA_Close','Upper','Lower']].plot(figsize=(15,5))
```

```python
df
```

```python
ShortEMA = df['Close'].ewm(span=12, adjust=False).mean()
LongEMA = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = ShortEMA - LongEMA
df['Signal_line'] = df['MACD'].ewm(span=9, adjust=False).mean()
```

```python
df
```

```python
df[['MACD','Signal_line']].plot(figsize=(15,5))
```

```python
def buy_sell(signal):
    Buy = []
    Sell = []
    flag = -1

    for i in range(0,len(signal)):
        if signal['MACD'].iloc[i] > signal['Signal_line'].iloc[i]:
            Sell.append(np.nan)
            if flag!=1:
                Buy.append(signal['Close'].iloc[i])
                flag = 1
            else:
                Buy.append(np.nan)
        elif signal['MACD'].iloc[i] < signal['Signal_line'].iloc[i]:
            Buy.append(np.nan)
            if flag!=0:
                Sell.append(signal['Close'].iloc[i])
                flag = 0
            else:
                Sell.append(np.nan)
        else:
            Buy.append(np.nan)
            Sell.append(np.nan)

    df['Buy'] = Buy
    df['Sell'] = Sell
```

```python
df = df.drop(['Buy_Signal_Price', 'Sell_Signal_Price'], axis=1)
```

```python
buy_sell(df)
```

```python
df
```

```python
plt.figure(figsize=(15,6.5))
plt.scatter(df.index, df['Buy'], color='green', label='Buy', marker='^', alpha=1)
plt.scatter(df.index, df['Sell'], color='red', label='Sell', marker='v', alpha=1)
plt.plot(df['Close'],label='Close Price',alpha =0.35)
plt.legend(loc='upper left')
plt.show()
```

```python
def calculate_stochastic_oscillator(df, k_period=14, d_period=3):
    # Calculate %K
    df['Lowest_Low'] = df['Low'].rolling(window=k_period).min()
    df['Highest_High'] = df['High'].rolling(window=k_period).max()
    df['%K'] = ((df['Close'] - df['Lowest_Low']) / (df['Highest_High'] - df['Lowest_Low'])) * 100

    # Calculate %D
    df['%D'] = df['%K'].rolling(window=d_period).mean()

calculate_stochastic_oscillator(df)
```

```python
df
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df['%K'], label='%K')
plt.plot(df['%D'], label='%D', linestyle='dashed')
plt.title('Stochastic Oscillator')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
def buy_sell_STOCH(df):
    Buy = []
    Sell = []
    flag = -1

    for i in range(0,len(df)):
        if df['%K'].iloc[i] > df['%D'].iloc[i] or df['%K'].iloc[i] < 20:
            Sell.append(np.nan)
            if flag!=1:
                Buy.append(df['Close'].iloc[i])
                flag = 1
            else:
                Buy.append(np.nan)
        elif df['%K'].iloc[i] < df['%D'].iloc[i] or df['%K'].iloc[i] > 80:
            Buy.append(np.nan)
            if flag!=0:
                Sell.append(df['Close'].iloc[i])
                flag = 0
            else:
                Sell.append(np.nan)
        else:
            Buy.append(np.nan)
            Sell.append(np.nan)

    df['Buy_Stoch'] = Buy
    df['Sell_Stoch'] = Sell

buy_sell_STOCH(df)
```

```python
def buy_sell_Bbands(df):
    Buy = []
    Sell = []
    flag = -1

    for i in range(0,len(df)):
        if df['Close'].iloc[i] < df['Lower'].iloc[i]:
            Sell.append(np.nan)
            if flag!=1:
                Buy.append(df['Close'].iloc[i])
                flag = 1
            else:
                Buy.append(np.nan)
        elif df['Close'].iloc[i] > df['Upper'].iloc[i]:
            Buy.append(np.nan)
            if flag!=0:
                Sell.append(df['Close'].iloc[i])
                flag = 0
            else:
                Sell.append(np.nan)
        else:
            Buy.append(np.nan)
            Sell.append(np.nan)

    df['Buy_Bbands'] = Buy
    df['Sell_Bbands'] = Sell

buy_sell_Bbands(df)
```
